## Exporting data to text file

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('quora_duplicate_questions.tsv', sep = '\t')
```

```python
df = df.dropna(how='any',axis=0) 
```

```python
df
```

```python
df.drop(labels=['id','qid1','qid2','is_duplicate'], axis = 1)
```

```python
que_df = pd.concat((df['question1'], df['question2']),axis =0,ignore_index=True)
```

```python
with open('data.txt', 'a+') as f:
  for item in que_df.values.tolist():
    f.write(item+ '\n')
```
